```python
# This Jungle Cruise game was written and coded by Tenbite Abebe

# Source: This movie was directed by Jaume Collet-Serra
# The movie was also written by Glenn Ficarra, John Requa, and Michael Green
# Below is the docstring that describes what is in this game 
"""
     Introduction: This game is based on the movie Jungle Cruise and it consists 
     of three stages. The game requires the player to think about which 
     moves to make to reach the final round where the player can get the chance to 
     get a mystical plant that heals everything. Follow the guided prompts to 
     learn more about the characters and the mysteries of the Amazon forest. 
 
     Level 1: Finding The Amazon Map
     Level 2: Crossing the Dangerous Amazon River 
     Level 3: Arriving at the Cave to Get the Mystical Flower

""" 
#The game begins here with the start game defined function
def start_game(): # Starting defined function that is also important in the end 
    print('---------------------------------------------------------------------')
    print("""Hello there, this cruise sails the Amazon and your Adventure starts here. 
            Lily Houghton and her brother Jack Houghton are physicians looking 
            for a map that can guide them through the Amazon. This map will help
            them to find a mystical tree called tears of the moon that possesses
            miraculous healing properties. The plant can stop diseases, viruses 
            and could bring someone back from the brink of death. This plant has 
            been hidden in the Amazon and only appears when there is a full moon.
    """)
    input(prompt ='It is going to be great, Press Enter')
    print('---------------------------------------------------------------------')
    # The input prompt & print allow a text to be displayed 
    # The player can then decide what they want to do next

    # The user defined variable for player name
    print ('Start your journey by introducing yourself to Lily, Frank, and Jack.')
    player_name = input(prompt = 'Enter Name Here ') # The player puts in a name 

    print('---------------------------------------------------------------------')
    
    print(f"""Welcome {player_name.capitalize()}""")
    # The capitilize function was used here to make sure they can input a name with those letters 
    # User defined variable for amazon location
   
    def level_1(): 
        # Start by defining level one to allow code organization 
        
        amazon_location = input(prompt = """Where is the Amazon Forest Located? 
                            1. Brazil 
                            2. South Africa
                            3. China
        """).lower() 
    # Put in the .lower function to make sure it looks for the typed in answers 
    # Developing the conditional statements 
    
        if "1" in amazon_location or "brazil" in amazon_location:
            print("You've Got it!\n")
            input(prompt = "Bravo, Press Enter\n")
            # The player can pick this option to move on to next question 

        elif "2" in amazon_location or "south africa" in amazon_location or "africa" in amazon_location:
            print("Wrong Place.\n")
            input(prompt = "Not there, Try Again\n")
           # This option will allow the player to try again because the answer is wrong 
        elif "3" in amazon_location or "china" in amazon_location:
            print("That's too far\n")
            input(prompt = "Try Again\n")
          # This option will allow the player to try again because the answer is wrong 
        else:
            print("Invalid Entry")
            input(prompt = "Incorrect, give it another try.") 
          # The player can try to choose an answer again 

    level_1() # Level one stage is here

    # The user has entered level One (Finding The Amazon Map)  
    # The player can use the choices given to answer the questions on each quest
    print('---------------------------------------------------------------------')

    def quest_1(): 
        map_1 = input(prompt = """
        You are in the city of London, while Jack is delivering a speech in the historical 
        medicine headquarters, your mission is to go to the artifacts office and retrieve 
        the original map of the Amazon forest. 

            How will you get inside the locked door?
            1) Pick the lock of the door 
            2) Go through the window 
            3) Sneak in using the back door
        """).lower()
    # Put in the .lower function to make sure it looks for the typed in answers
    # Developing the conditional statements for quest_1
    
        if "1" in map_1 or "pick lock" in map_1 or "pick door" in map_1:
            print("You've opened the door!\n")
            input(prompt ="Let's Keep Going, Press Enter\n")
            # This option allows player to go to the next quest

        elif "2" in map_1 or "go through" in map_1 or "window" in map_1:
            print("Now you have entered the office, be careful not to break anything.\n")
            input(prompt ="Let's keep Going, Press Enter\n")
           # This option allows player to go to the next quest
        elif "3" in map_1 or "sneak in" in map_1:
            print("Good job sneaking in, be careful! Don't make any noise.\n")
            input(prompt ="Press Enter for Next Task\n")
       # Options here from 1 to 3 the player can pick from and be sent to the next level 
       
        else:
            print("Invalid Entry")
            input(prompt = "Give it another try.")
       # If the player doesnt put in a valid answer, they will be sent back to try again 
            
    quest_1() # The first question here is finished 

    print('---------------------------------------------------------------------')

    def quest_2():
        #next user defined variable for finding map
        map_2 = input(prompt = """
        Lily has informed you that the map should be in this room, decide where you want 
        to look.   

            Hint: She mentioned all the prehistoric objects have been put out for display.
            1) In the safe 
            2) On the center table
            3) The box under the shelves
    """).lower()
    # Put in the .lower function to make sure it looks for the typed in answers
    # Developing the conditional statements for quest_2
    
        if "1" in map_2 or "open safe" in map_2 or "safe" in map_2:
            print("Not in there!\n")
            input(prompt = "Pick another try\n")
        # The player can get another try since this answer is not correct 

        elif "2" in map_2 or "go through" in map_2 or "window" in map_2:
            print(" You have found the greatest map of all time!\n")
            input(prompt = "Press enter to continue\n")
        # This answer is the correct one that will help the player move on 
        
        elif "3" in map_2 or "in box" in map_2 or "under shelves" in map_2:
            print("Oh no it is not there.\n")
            input(prompt = "Pick another try\n") 
        # The player can get another try since this answer is not correct 
        else:
            print("Invalid Entry")
            input(prompt = "You have not made the right option, give it another try.")
            # The player has another shot to try again 
    quest_2()

    print('---------------------------------------------------------------------')

    # The player has aquired the map, now moves on to level two
    # This is LEVEL TWO: Crossintg the Dangerous Amazon River

    def level_2():
        print(""" You've got the map. In level two. Lily, Jack, and you have gone to
        brazil and you will need to cross the Amazon river to continue! Be safe and be 
        cautious. Don't forget, it is the wild!
        """)
        input(prompt = "Let us push through, Press Enter")
    level_2()  
    # Indicating that this is the second level  

    # Here the player will try and guess the correct amount of gold coins to pay through loops
    # Declaring an object
    number = (1,5) 
    guesses = 10

    # Looping until the player can pick the correct bargain for Frank to accept
    while True > 0:
        amazon_1 = input(prompt = """
        Frank is the captain of the ship and has agreed to sail you along the 
        Amazon to your quest in finding the mystical tree. Now you have to make a 
        deal to pay Frank for his service. You currently have 5 gold coins, bargain (1-5) 
        How much will Frank accept to take you down the amazon river? 
        """)

        # Converting choice for numberical comparison 
        amazon_1 = int(amazon_1)

        # Converting statement based on user input 
        if amazon_1 == 3:
            print("""He accepts your offer and added 1 gold coins for the shipâ€™s fuel, 
                  you can give this amount to Frank after safely getting back. 
                  """) 
        # This if statement is the correct one the player should try and bargain for
            break

        elif amazon_1 != number:

            if amazon_1 > 3:
                print("Too much might need to keep some for yourself.")
                # If the player picks a number bellow 3 they will have to try again

            elif amazon_1 < 3:
                print("Frank said it's too low!")
                # If the player picks a number above 3 they will have to try again
            else:
                print('Error, incorrect bargain')
    # These other statements (elif, if and else that is not 3) will all allow the player another guess 
    print('---------------------------------------------------------------------')

    def quest_4():
    # Next user defined varibale for going through the amazon
        amazon_2 = input(prompt = """
        Now that you have a ship, what time would you want to leave for your quest? 
        If you leave at dawn, the boat may be taken by the rich investor in town that 
        owns the boat. However, if you leave now you can also avoid the high tides. 
        Pick your choice. 
            1. Leave now
            2. Leave at dawn 
    """).lower()
    # Put in the .lower function to make sure it looks for the typed in answers
    # Developing the conditional statement for quest_4
    
        if "1" in  amazon_2 or "leave now" in  amazon_2 or "leave" in amazon_2 or "now" in amazon_2:
            print("""Good work! Now you have started your journey, you have been on 
            the ship for 3 hours.  
            """) 
            input(prompt = "Press Enter to keep sailing")

        elif "2" in  amazon_2 or "leave at dawn" in  amazon_2 or "dawn" in  amazon_2:
            print("""It is 6 am, and you are ready to leave for your journey 
            to the Amazon. Nilo Nemolato, the owner of the ship, has come to take 
            it from Frank. You have to quickly pack and board before he catches you.
            """)
            input(prompt = "Press enter to release franks cheetah in the street to create a distraction")
        # These two choices of if and elif will take the player to the next levels 
        else:
            print(prompt = "Invalid Entry")
            input("Try Again")
        # This else clause will help player redo the question if the put something that is invalid 
    quest_4() # Question 4 has ended here moving on to the next quest 
    
    def exit(): # If the player wants to exit the game this will help them pick
        print(""" You have failed this game""")
        print(""" Game Over""")

        choice = input(prompt = """
         Do you want to play again? (yes/no)""")

        if 'yes' in choice:
            print("""Glad you enjoyed the game!""")
            start_game() # Calling this statement will restart the game 
        
        elif 'no' in choice:
            print("""Thank you for playing!""")
            exit() # Calling this stataemnt will exit player off this game
        
        else: # Relaunch the game 
            print(""" Please reply with "yes" or "no" """)

    def fail(): # If the player fails a question they are out of the game
        print(""" You have failed this game""")
        print(""" Game Over""") # Here it will print game over for the player

        choice = input(prompt = """ Do you want to play again? (yes/no)
        """)

        if 'yes' in choice:
            print("""Glad you enjoyed the game!""")
            start_game()
            # This start_game() will take the player back to the beginning 
        elif 'no' in choice:
            print("""Thank you for playing!""")
            exit()
        else: # Relaunch the game 
            print(""" Please reply with "yes" or "no" """)
            fail() # This function is needed in order to make the player get out of the game 
    print('---------------------------------------------------------------------')
   
    def quest_5():
        import random 
        # Developing a question list that has questions to be randomly selected reach time a play runs game
        question_list = [
    """\nSailing on the water for hours can make you hungry and tired, do you want 
    to have some food? """,
    """\nYou have become exhausted with moving your bags onto the ship, do you want 
    to get some rest? """,
    """\nYou can do some fishing with Frank to feed your brother, do you want to catch 
    your dinner? """,
    """\nYou should sit and relax to get a break when sailing across the Amazon river,
    do you want to have some dinner? """, ]
        
        # Creating a randomized question selection from the list 
        question = random.choice(question_list)
        print(question)
        print(" ")
            
        answer = input(prompt = " (Yes/No) > ")
            
        # Changing the object to lowercase by using .casefold()
        answer = answer.casefold()
            
        if answer == "yes" or "y" in answer:
            print(""" It is good to eat and rest, now you will have all the energy.
            """) 
            # The player is able to go to next question with this option 
        elif answer == "no" or "n" in answer:
            print(""" You will be too exhausted to continue this trip.
            """) 
            fail() # The player will lose the game due to no food and rest 
        else:
            print("Invalid, Try Again")
            # The player can try to input answer again 
         
    quest_5()
        
    print('---------------------------------------------------------------------')

    def quest_6():
        # Next user defined varibale for going through the amazon
        amazon_3 = input(prompt = """
        The indigenous people of the forest feel like you are intruding 
        into their home and all of a sudden they have started attacking your ship. 
          What should you do?
            1. Let Frank speak to them in their language to make them feel safe 
            2. Attack them and try and escape 
    """).lower()
    # Put in the .lower function to make sure it looks for the typed in answers
    # Developing the conditional statement for quest_6
    
        if "1" in  amazon_3 or "speak to them" in  amazon_3 or "talk" in amazon_3 or "SPEAK" in amazon_3 or "safe" in amazon_3:
            print("""You have befriended the indigenous people and they will help 
            you reach the cave where the mystical tree is located. You have moved 
            on to the next level.   
            """) 
            input(prompt = "Press Enter to keep continue ")
            # The player picked the correct answer and can move onto the next round

        elif "2" in amazon_3 or "attack" in amazon_3 or "kill" in amazon_3:
            print("""You have attacked the indigenous people, however, they outnumbered 
            you and fought back and imprisoned you in their forest.
            """)
            fail() 
            # The way a player fails this game is by attacking the indigenous people to restart game 
        else:
            print("Invalid Entry")
            input(prompt = "Try Again")
            # The player can try to input answer again 

    quest_6()
    print('---------------------------------------------------------------------')
    # This print above allows there to be a line break between questions 

    # The player has entered LEVEL 3: the last level
    def level_3():
        print(""" You've made it through the Amazon river. In level three. Time to
        find the mystical plant.
        """)
        input(prompt = "Keep your eyes open, Press Enter")
        # Allows player to press enter to continue the game 
    level_3()

    def quest_7():
    # Next user defined varibale for the cave level
        cave_1 = input(prompt = """
        You have reached the cave. Somewhere in this cave hidden behind the trees and 
        protected by the wolves is the mystical tree. Which way do you want to head to 
        reach this magical tree?
            1. Walk up the pyramid hill 
            2. Go through the tunnel
    """).lower()
    # Put in the .lower function to make sure it looks for the typed in answers
    # Developing the conditional statement for quest_7
        if "1" in cave_1 or "walk up" in  cave_1 or "walk hill" in cave_1 or "pyramid" in cave_1:
            print("""Great job you have gone up the hill, beware of the snakes. 
            """) 
            input(prompt ="Let's find that flower, Press enter")
            # This allows the player to move into next quest 

        elif "2" in  cave_1 or "go through tunnel" in  cave_1 or "tunnel" in  cave_1:
            print("""
            """)
            input(prompt = "Your torch is on to scare the bats, Press Enter")
            # This allows the player to move into next quest 
            # The text will be printed out to allow player to press enter 
        else:
            print("Invalid Entry")
            input(prompt = "Try Again")
            # The player can try to input answer again 
    quest_7()

    print('---------------------------------------------------------------------')
    # This print above allows there to be a line break between questions 
    
    def quest_8():
        # Player is almost at the end of game
        cave_2 = input(prompt = """
       The full moon is disappearing and the tree will be gone soon. You have to 
       get to the one flower that is left before all the leaves fall. How should you get there? 
            1. Climb up the tree bark 
            2. Swing from the top of the pyramid to the closest branch 
    """).lower()
    # Put in the .lower function to make sure it looks for the typed in answers 
    # Developing the conditional statement for quest_8 
    
        if "1" in cave_2 or "climb the tree" in  cave_2 or "tree bark" in cave_2 or "climb" in cave_2 or "tree" in cave_2:
            print("""Press Enter for speed
            """) 
            input(prompt = "The cave is shaking, Press enter") 
            # The player can keep going with this option 

        elif "2" in  cave_2 or "swing" in  cave_2 or "swing from pyramid" in  cave_2:
            print("""Hold Tight, Press Enter
            """)
            input(prompt = "Great, almost there, Press Enter")
           # The player can keep going with this option 
        else:
            print("Invalid Entry") 
            input(prompt = "Try Again")
            # The player can try to input answer again 
    quest_8()

    print('---------------------------------------------------------------------')
   
            
    def winner(): #The winner is definied to guide player to finish the game
        print("""The Amazon is a dangerous forest but you have made it out!  
        """)
        choice = input (prompt = """ Do you want to play again? (yes/no)""")

        if 'yes' in choice:
            print(""" I'm glad you liked it!""")
            start_game() 
            # Here the game will be restated when calling start_game()
        elif 'no' in choice:
            print(""" Thank you for playing!""")
            exit() #The player will be taken out of the game
        else: # Relaunch this game
            print(""" Please only with "yes" or "no." """)
            
    def quest_9():
    # The player is at the end of game and has to pick the last decision
        
        cave_3 = input(prompt = """
       Great you have acquired the flower! However, Frank has been caught by the 
       traps made for invaders of the cave and is dying. How will you use the medicinal 
       power that can heal anything?
           1. Take it home to save the world 
           2. Save Frank's Life 

    """).lower()
    # Put in the .lower function to make sure it looks for the typed in answers 
    # Developing the conditional statement for quest_9
    
        if "1" in cave_3 or "take it home" in  cave_3 or "save the world" in cave_3 or "1" in cave_3 or "Save the world" in cave_3:
            print(""" CONGRATULATIONS! You have won the game great choice! Saving the world is smart 
            because the flower will allow Lily to create medication from it that can end
            all diseases! Thank you for playing. 
            """) 
            winner() # Taking player to defined winner text 

        elif "2" in  cave_3 or "save frank's life" in  cave_3 or "save life" in  cave_3:
            print(""" WINNER! You saved your friend. That was a good call. Good friends are hard to come by 
            therefore, when you get them, be sure to protect them.
            """)
            winner() # Print this text for the winner of the game 
        else:
            print("Invalid Entry")
            input(prompt = "Try Again")
    quest_9() # The game questions have ended, close off the quest_9 

start_game() # Here the game will be closed so that the questions will come in the order there were coded
```
